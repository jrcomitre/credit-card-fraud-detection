```python
# Importando bibliotecas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configurações de estilo dos gráficos
sns.set(style="whitegrid")
%matplotlib inline
```

```python
# Carregando o dataset
df = pd.read_csv(r'C:\Users\user\Documents\credit-card-fraud-detection\data\creditcard.csv')
df.shape
```

```python
# Inspeção inicial dos dados (Exibindo as primeiras linhas do dataset)
df.head()
```

```python
# Verificando valores ausentes (null)
df.isnull().sum()
```

```python
# Tipos de dados e estrutura
df.info()
```

```python
# Distribuição da variável alvo (Class)
class_counts = df['Class'].value_counts()
plt.figure(figsize=(6,4))
sns.barplot(x=class_counts.index, y=class_counts.values, palette=["blue", "red"])
plt.title('Distribuição das Classes (0 = Não Fraude, 1 = Fraude)')
plt.xlabel('Classe')
plt.ylabel('Quantidade')
plt.xticks([0, 1], ['Não Fraude', 'Fraude'])
plt.tight_layout()
plt.show()
```

```python
# Estatísticas do valor das transações por classe
df.groupby("Class")["Amount"].describe()
```
